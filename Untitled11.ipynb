```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re, unicodedata, difflib

FILE = "NUM25-F7.xlsx"

# 1) Lire brut pour détecter la bonne ligne d'en-tête
raw = pd.read_excel(FILE, header=None)

def normalize_text(s):
    if pd.isna(s): return ""
    s = str(s).replace("’","'")
    s = unicodedata.normalize("NFKD", s)
    s = "".join(ch for ch in s if not unicodedata.combining(ch))
    return s

# Cherche la 1re ligne qui contient "Temps quotidien" (dans n'importe quelle colonne)
mask_header_rows = raw.apply(lambda r: r.astype(str).map(normalize_text)
                                        .str.contains("Temps quotidien.*exposition.*ecrans", case=False, regex=True, na=False)).any(axis=1)

if not mask_header_rows.any():
    raise ValueError("Je n'ai pas trouvé la ligne d'en-tête. Vérifie que le fichier contient bien la ligne avec 'Temps quotidien d’exposition aux écrans'.")

header_row = mask_header_rows[mask_header_rows].index[0]
print(f"Ligne d'en-tête détectée : {header_row}")

# 2) Recharger avec la bonne ligne d'en-tête
df = pd.read_excel(FILE, header=header_row)

# 3) Nettoyage des noms de colonnes
def clean_cols(cols):
    out = []
    for c in cols:
        c = "" if pd.isna(c) else str(c)
        c = normalize_text(c).strip()
        out.append(c)
    return out

df.columns = clean_cols(df.columns)

# 4) Supprimer les colonnes vides / 'Unnamed'
df = df.loc[:, ~df.columns.str.match(r"^Unnamed", na=False)]
df = df.loc[:, df.columns.str.len() > 0]

print("Colonnes après nettoyage :", list(df.columns))

# 5) Trouver les colonnes utiles même si le nom varie un peu
def best_match(targets, columns, cutoff=0.6):
    cols_norm = {normalize_text(c).lower().strip(): c for c in columns}
    for t in targets:
        key = normalize_text(t).lower().strip()
        if key in cols_norm:
            return cols_norm[key]
    # sinon on prend le plus proche
    candidates = list(cols_norm.keys())
    close = difflib.get_close_matches(normalize_text(targets[0]).lower().strip(), candidates, n=1, cutoff=cutoff)
    return cols_norm[close[0]] if close else None

col_time = best_match(
    ["Temps quotidien d'exposition aux ecrans", "Temps quotidien d’exposition aux ecrans", "Temps quotidien"],
    df.columns
)
col_ens  = best_match(["Ensemble", "Total", "Population totale"], df.columns)
col_f    = best_match(["Femmes"], df.columns)
col_h    = best_match(["Hommes"], df.columns)

if not col_time or not col_ens:
    raise ValueError(
        "Impossible de repérer les colonnes clés.\n"
        f"Trouvé -> temps: {col_time}, ensemble: {col_ens}\n"
        f"Colonnes dispo: {list(df.columns)}"
    )

# 6) Garder les lignes pertinentes (catégories de temps non vides)
df = df[df[col_time].notna()].copy()

# 7) Convertir les pourcentages (gère les virgules, % et espaces)
def to_num(x):
    if pd.isna(x): return np.nan
    if isinstance(x, (int, float)): return x
    s = str(x).replace(",", ".")
    s = re.sub(r"[^0-9.\-eE]", "", s)
    return pd.to_numeric(s, errors="coerce")

for c in [col_ens, col_f, col_h]:
    if c and c in df.columns:
        df[c] = df[c].apply(to_num)

# 8) Graphique Ensemble
plt.figure(figsize=(9,5))
plt.bar(df[col_time].astype(str), df[col_ens])
plt.xlabel("Temps quotidien d'exposition aux écrans")
plt.ylabel("Ensemble (%)")
plt.title("Temps d'exposition aux écrans – Ensemble")
plt.tight_layout()
plt.show()

# 9) Graphique Hommes vs Femmes si dispo
if col_f and col_h and col_f in df.columns and col_h in df.columns:
    idx = np.arange(len(df))
    w = 0.4
    plt.figure(figsize=(9,5))
    plt.bar(idx - w/2, df[col_f], width=w, label="Femmes", alpha=0.85)
    plt.bar(idx + w/2, df[col_h], width=w, label="Hommes", alpha=0.85)
    plt.xticks(idx, df[col_time].astype(str))
    plt.xlabel("Temps quotidien d'exposition aux écrans")
    plt.ylabel("Pourcentage")
    plt.title("Temps d'exposition aux écrans – Hommes vs Femmes")
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("Comparatif Hommes/Femmes ignoré (colonnes non trouvées).")

# 10) Sauvegarde du tableau nettoyé
df.to_csv("NUM25-F7_nettoye.csv", index=False)
print("Enregistré : NUM25-F7_nettoye.csv")
```
